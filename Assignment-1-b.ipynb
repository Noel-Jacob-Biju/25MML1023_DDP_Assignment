Write a PySpark program to:

- Create a DataFrame with at least 5 records (columns: ID, Name, Age, Salary)

- Display the schema

- Show all records

- Filter employees with salary greater than 30,000

```python
from pyspark.sql import SparkSession

# Initialize Spark Session
spark = SparkSession.builder.appName("Assignment1").getOrCreate()
```

```python
# 1. Create a DataFrame with 5 records
data = [
    (1, "Alice", 28, 45000),
    (2, "Bob", 35, 28000),
    (3, "Charlie", 30, 52000),
    (4, "David", 22, 15000),
    (5, "Eve", 40, 60000)
]
columns = ["ID", "Name", "Age", "Salary"]
df = spark.createDataFrame(data, schema=columns)
```

```python
# 2. Display the schema
df.printSchema()
```

```python
# 3. Show all records
df.show()
```

```python
# 4. Filter employees with salary > 30,000
high_earners = df.filter(df.Salary > 30000)
high_earners.show()
```
